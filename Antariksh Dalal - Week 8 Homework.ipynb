#### Find the signals and create a new column that categorizes the incidents.

There is no "right" answer here, just a way for you as a emerging data scientist to think about how to parse data and categorize it. Just declare categories and justify your logic! I am as interested in your reasoning, as I am in your code!

```python
#Your code here:

import pandas as pd

df = pd.read_csv("RAND_Database_of_Worldwide_Terrorism_Incidents.csv", encoding='latin1')
#df.head()

# Dictionary of keywords associated with each category (economic, political, and religious)
keywords = {
    'economic': ['money', 'financial', 'economy', 'wealth', 'poverty', 'resources'],
    'political': ['government', 'regime', 'election', 'democracy', 'liberation', 'independence'],
    'religious': ['god', 'faith', 'holy', 'sacred', 'religion', 'church', 'mosque', 'temple']
}

def categorize_description(description):
    """Categorize a given description into economic, political, religious, or a combination of these categories based on the above pre-defined category and keyword associatiosn"""
    if isinstance(description, str):
        # Convert description to lowercase for case-insensitive matching
        description = description.lower()
        categories = []
        # Check for keywords from each category in the description
        for cat, keys in keywords.items():
            if any(key in description for key in keys):
                categories.append(cat)
        # Join found categories with '+' or return 'uncategorized' if none found
        return '+'.join(categories) if categories else 'uncategorized'
    return 'uncategorized'

# Creating a new 'Category' column
df['Category'] = df['Description'].apply(categorize_description)

# Create a new DataFrame with only categorized incidents
categorized_df = df[df['Category'] != 'uncategorized']
print(categorized_df[['Description', 'Category']].head(10))

print("\nCategory Distribution:")
print(df['Category'].value_counts())

df.tail()
categorized_df.tail()
```
